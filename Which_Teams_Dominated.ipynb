```python
import pandas as pd
```

```python
results = pd.read_csv("results.csv")
```

```python
home_matches = results[results['neutral'] == False]
```

```python
home_goals = home_matches.groupby('home_team')['home_score'].mean().reset_index()
```

```python
home_conceded = home_matches.groupby('home_team')['away_score'].mean().reset_index()
```

```python
home_performance = pd.merge(home_goals, home_conceded, on='home_team', suffixes=('_scored', '_conceded'))
```

```python
home_performance.head()
```

```python
away_matches = results[results['neutral'] == False]
```

```python
away_goals = away_matches.groupby('away_team')['away_score'].mean().reset_index()
```

```python
away_conceded = away_matches.groupby('away_team')['home_score'].mean().reset_index()
```

```python
away_performance = pd.merge(away_goals, away_conceded, on='away_team', suffixes=('_scored', '_conceded'))
```

```python
away_performance.head()
```

```python
home_performance['goal_difference'] = home_performance['home_score'] - home_performance['away_score']
```

```python
away_performance['goal_difference'] = away_performance['away_score'] - away_performance['home_score']
```

```python
home_dominance = home_performance.sort_values(by='goal_difference', ascending=False)
```

```python
away_dominance = away_performance.sort_values(by='goal_difference', ascending=False)
```

```python
print("Teams Dominating at Home:")
print(home_dominance.head())
```

```python
print("Teams Dominating Away:")
print(away_dominance.head())
```
