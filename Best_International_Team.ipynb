```python
import pandas as pd
```

```python
results = pd.read_csv("/content/sample_data/results.csv")
```

```python
results.head()
```

```python
non_friendlies = results[results['tournament'] != 'Friendly']
```

```python
team_performance = non_friendlies.groupby('home_team').agg({
    'home_score': 'sum',
    'away_score': 'sum'
}).reset_index()
```

```python
team_performance.columns = ['team', 'home_goals', 'away_goals']
```

```python
team_performance['total_goals'] = team_performance['home_goals'] + team_performance['away_goals']
```

```python
team_performance.head()
```

```python
team_performance['goal_difference'] = team_performance['home_goals'] - team_performance['away_goals']
```

```python
best_teams = team_performance.sort_values(by=['total_goals', 'goal_difference'], ascending=False)
```

```python
best_teams.head()
```

```python
print("Best Team of All Time:", best_teams.iloc[0]['team'])
```
